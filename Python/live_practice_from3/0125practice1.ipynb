### Index로 접근하여 풀이하기

```python
for element in range(len(list)):
```

````python
# 아래에 코드를 작성하세요.
def sum_list_index(numbers):
    l =[]
    for i in range(len(numbers)):
        lst = numbers[i]
        for j in range(len(lst)):
            l.append(lst[j])
    return sum(l)
            
````

````python
print(sum_list_index([[1, 4], [10, 5], [20, 30]]))
````

### while 문을 활용하여 풀이하기

```python
while ____:
```

````python
# ## 아래에 코드를 작성하세요.
def sum_list_while(numbers):
    cnt = 0
    t = 0
    while cnt < len(numbers):
        t += sum(numbers[cnt])
        cnt += 1
    return t
        
    
    ##
````

````python
print(sum_list_while([[1, 4], [10, 5], [20, 30]]))
````

### 학생별 출력

* 아래의 리스트를 반복하여 **학생별 총합**을 순서대로 `출력`하시오.

* `sum` 함수 사용 금지

````python
students = [
 [100, 80, 100],
 [90, 90, 60],
 [80, 80, 80]
]
````

````python
students = [
 [100, 80, 100],
 [90, 90, 60],
 [80, 80, 80]
]
````

````python
# 아래에 풀이를 작성하세요. 함수 정의는 필요없습니다.

b = zip(*students)  # 이것이 이중 리스트이기 때문에 벗기고 난 다음에 그다음에 zip을 해주는 것이다

for i in b:
    total = 0
    for j in i:
        total += j
    print(total)
    
    
````
