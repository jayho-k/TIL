# 09.데이터 선택(조건)
조건에 해당하는 데이터 선택
더 공부하고 싶다면 구글에 pandas string 이라고 치면 된다.

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df['키'] >= 185
```

```python
filt = (df['키'] >= 185)
df[filt]
# 키가 185이상만
```

```python
df[~filt]
# 물결을 하면 반대 값이 나옴 not
```

```python
# df.loc[row, col]

df.loc[filt, '수학']
# 키가 185인 학생들의 수학성적
```

```python
df.loc[filt, ['수학', '이름', '과학']]
```

## 다양한 조건

### 그리고 (&)

```python
filt2 = df['학교'] == '북산고'
filt2
```

```python
df.loc[filt & filt2,['이름','수학', '과학']]
```

### 또는 (|)

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)]
```

### str함수

```python
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
filt = df['이름'].str.contains('송') # 이름에 송이 들어가는 사람
df[filt]
```

```python
df[~filt] # 이름에 송이 안들어가는 사람
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) # 특기가 파이썬이거나 자바인사람
df[filt]
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)   # 특기가 파이썬이거나 자바인사람  소문자 대문자 다포함
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java', na = False )
df[filt]
# Nan데이터가 있으면 처리를 하지 못한다.
# 따라서 na = False를 써주어야 한다.
```
