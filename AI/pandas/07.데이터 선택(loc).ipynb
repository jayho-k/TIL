# 7. 데이터 선택 (loc)
df.loc[인뎃스, 칼럼]

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df.loc['1번'] # 인덱스 1번에 해당하는 전체 데이터
```

```python
df.loc['1번', '국어']
```

```python
df.loc[['1번', '2번'],[ '영어']]
```

```python
df.loc[['1번', '2번'],[ '영어', '국어']]
```

```python
df.loc['1번':'5번', '국어': '사회']
```
