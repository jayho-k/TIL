##  데이터에 함수 적용(apply)


```python
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) #키 데이터
df
```

```python
def capitalize(lang):
    if pd.notnull(lang):  #Nan이 아닌지 확인
        return lang.capitalize()
    
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

```python
# 똑같은 결과가 나온다.
df['SW특기'] = df['SW특기'].str.capitalize()
df
```
